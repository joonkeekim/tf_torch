keyword parameter
python은 파라미터의 이름을 명시해서 전달 가능
안쓰면 뭐 순서대로 주는거지

```
def multiply(x,y):
  if x>10:
    return (x)*y
  return (x +2) * y
multiply(1,5)
```

```
def add_mul(x,y):
  s = x*y
  m = x+y
  return s,m
c = add_mul(20,3)
print(type(c))
print(c)
```

여러개로 반환받으면 이렇게 tuple로 반환을 받는다

parameter를 만들 때 가변길이 parameter로 만들 수 있다.

1.   *하나 붙으면 tuple
2.   **붙으면 딕셔너리






```
def test(*x): #args를 많이 쓰자. arguments
  for item in x:
    print(item)
test(1,2,3,4,5,6)
```

```
def test2(**kwargs):
  for key, value in kwargs.items():
    print('key', key , 'value', value)
  #print(type(kwargs))
test2(a=1,b=2,c=3,d=4,name = 'Bob')
```
