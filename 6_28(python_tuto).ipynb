```
class Person:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def eat(self,food):
    print('{}는 {}를 먹습니다'.format(self.name,food))
  def sleep(self,minute):
    print('{}는 {}분동안 잡니다'.format(self.name,minute))
  def work(self,minute):
    print('{}는 {}분동안 일할 준비를 합니다'.format(self.name,minute))#여기 바뀜
class Student(Person): #상속하는 방법
  def __init__(self,name,age):
    self.name = name
    self.age=age
  def work(self,minute):
    super().work(20)
    print('{}는 {}분동안 공부합니다'.format(self.name,minute))
class Employee(Person):
  def __init__(self,name,age):
    self.name = name
    self.age=age
  def work(self,minute):
    super().work(20)#super를 통해서 부모의 기능을 사용가능
    print('{}는 {}분동안 업무합니다'.format(self.name,minute))  
  
bob = Student('Bob',25)
bob.eat('BBq')
bob.sleep(30)
bob.work(60)
```

```
#point
#2차원 좌표평면 각 점(x,y)

class Point:
  def __init__(self,x,y):
    self.x = x
    self.y = y
  def print_pt(self):
    print('({},{})'.format(self.x,self.y))
  def __str__(self):
    return '({},{})'.format(self.x,self.y)
  def add(self,pt):
    new_x = self.x + pt.x
    new_y = self.y +pt.y
    return Point(new_x,new_y)
  def __add__(self,pt):
    new_x = self.x + pt.x
    new_y = self.y +pt.y
    return Point(new_x,new_y)
  def __sub__(self,pt):
    new_x = self.x - pt.x
    new_y = self.y -pt.y
    return Point(new_x,new_y)
  def __mul__(self,factor):
    return Point(self.x*factor,self.y*factor)  
  def __len__(self):
    return self.x **2 +self.y**2
  def get_x(self):
    return self.x
  def get_y(self):
    return self.y
  def __getitem__(self,index):
    if index ==0:
      return self.x
    elif index ==1 :
      return self.y
    else : 
      return self.x

p1 = Point(3,4)
p2 = Point(2,7)

p1.print_pt()#이렇게말고 그냥 print는
p2.print_pt()
#p3 = p1.add(p2)
p3 = p1+p2
p4 = p1-p2
p5 = p1*3
print(p1)
print(p2)
print(p3)
print(p4)
print(p5)
print(len(p5))
print(p5[0])
```

\d 숫자 캐릭터와 일치

^ 시작, $ 끝 각각 문자열의 시작과 끝을 의미

[]내부의 메타 캐릭터는 캐릭터 자체를 나타냄

[abck]는 a or b or c or k

-을 쓰면 해당 문자 범위에 속하는 문자하나

[a-d] a에서 d 사이

[0-9] 모든 숫자

[a-z] 모든 소문자

[A-Z] 모든 대문자

[a-zA-Z0-9] 모든 알파벳 수자

^이 맨앞에 사용되면 해당문자패턴이 아닌것들만 골라서

[^0-9]숫자빼고 다 not을 의미

```
re.search(r'[cbm]at','cat') #cat or bat or mat
```

```
re.search(r'[0-9]haha','1hahaha')#yes
re.search(r'[0-4]hahaha','7hahaha')#none
```

```
re.search(r'[abc.^]aron','.aron')
```

```
re.search(r'[^abc]aron','aaron')#이러면 none
re.search(r'[^abc]aron','#aron')#이건 있지 ^의 범위에 포함이 안 되니까
```

##반복적인 패턴##

'+' 1번이상의 패턴발생

'*' 0번이상

'?' 0혹은1번의 패턴발생

가능한 많은 패턴을 찾는다

```
re.search(r'a[bcd]*b','abcddccb')#a로 시작해서 [bcd]가 0번이상 반복하고 b로 끝나는 패턴
#후보는 세개
#ab, abcdb, abcddccb 이지만 가장 많은 패턴을 출력
```

```
re.search(r'b\w+a','banana')#b로 시작해서 알파벳과 숫자가 1번이상 반복되고 a로 끝나는 패턴
```

gropuing

()를 사용

매칭결과를 그룹별로 분리가능

패턴명시할때 각그룹을 괄호안에 넣어 분리하여 사용

```
m = re.search(r'\w+@.+','test@gmail.com')#모든문자가 한번이상 @있고 .이한번이상
m.group()
```

```
m = re.search(r'(\w+)@(.+)','test@gmail.com')#이렇게 하면 그룹별로 구분가능
print(m.group(1))
print(m.group(2))
print(m.group(0))#0은 기본값
```

```
import requests
from bs4 import BeautifulSoup

def get_news_content(url):
  response = requests.get(url)
  content = response.text

  soup = BeautifulSoup(content, 'html5lib')

  div = soup.find('div',attrs = {'id':'harmonyContainer'})

  content = ''
  for paragraph in div.find_all('p'):
    content += paragraph.get_text()

  return content

news1 = get_news_content('https://news.v.daum.net/v/20190617073049838')
print(news1)
```
