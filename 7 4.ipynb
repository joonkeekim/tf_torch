#시각화 graph visualization

load packages

```
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline #주피터 내부에 띄우겠다.
```

```
data = np.random.randn(50).cumsum() #주어진 축에 따라 값을 더해둔다.
data
```

```
plt.plot(data)
plt.show()
```

```
plt.subplot(1,2,1)#1행 2열 1번째
plt.subplot(1,2,2)#1행 2열 2번째
plt.show()
```

```
#multi graph
hist_data = np.random.randn(100)
scat_data = np.arange(30)
hist_data
```

```
plt.subplot(2,2,1)
plt.plot(data)#x축은 index y축은 값

plt.subplot(2,2,2)
plt.hist(hist_data, bins = 20)#y축 최대값?

plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) + 3*np.random.randn(30))

plt.show()
```

```
plt.plot(data, 'go')#색깔 + 마커모양으로 표현할 수 있다.

plt.show()
```

```
plt.figure(figsize=(10,5))#크기 조절 인치 단위임!!
plt.plot(data, 'k+')
plt.show()
```

```
plt.figure(figsize = (10,10))
#figure는 맨 앞에 있어야 한다!!!!!
plt.subplot(2,2,1)
plt.plot(data)#x축은 index y축은 값

plt.subplot(2,2,2)
plt.hist(hist_data, bins = 20)#y축 최대값?

plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) + 3*np.random.randn(30))

plt.show()
```

```
data = np.random.randn(30).cumsum()#누적합 뒤로갈 수록

plt.plot(data, 'k--', label = 'Default')
plt.plot(data, 'k-' , drawstyle ='steps-post')#step으로 주는거임
plt.legend()
#subplot을 안해서 겹쳐서 나오게 된다.
plt.show()
```

```
plt.plot(np.random.randn(1000).cumsum())
plt.title('ddd')
plt.xlabel('dd')
plt.show()
```

```
plt.plot(np.random.randn(1000).cumsum(),'k',label ='one')
plt.plot(np.random.randn(1000).cumsum(),'b.',label ='two')
plt.plot(np.random.randn(1000).cumsum(),'r-',label ='three')
plt.legend()
plt.show()#drag and drop 가능
```
