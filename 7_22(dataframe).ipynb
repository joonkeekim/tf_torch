```
def get_class_name(path):
  fname = os.path.basename(path)
  cls_name = fname.replace('.png','').split('_')[-1]#class 네임을 가져옴
  return cls_name
```

```
cls_name = get_class_name(path)
path, cls_name
```

#class 수 확인

```
class_name = []
for path in train_path:
  cls_name = get_class_name(path)
  class_name.append(cls_name)
```

```
len(class_name)#다 뽑아서 리스트에 넣음
```

```
class_name = [get_class_name(path) for path in train_path]#list compression
```

```
class_name[:10]
```

```
#unique한 label만 보고싶다면
uniques = np.unique(class_name, return_counts=True)#몇 개씩 있는지도 보여준다.
```

```
#plt.bar(uniques[0],uniques[1])
plt.bar(*uniques)#이렇게 하면 편하게 보이는거다.
plt.show()
```
